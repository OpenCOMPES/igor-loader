```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import ipywidgets as widgets
from ipywidgets import interact
from igor2 import binarywave
from pathlib import Path
from igor_loader import static_analysis
```

```python
#Load class
sp=static_analysis() 
```

```python
datapath="test_data/"
run=11
name="WSe2_"
```

```python
sp.load(datapath,name,run)
# dims=['Energy','Angular'] 
# sp.edit() #If you need to change the name of the dimensions
```

```python
# sp.data.plot() #Visualize the data
sp.edc() #check the edcs
```

Axis Scan (Fixed and Sweep)

```python

datapath="test_data/"
run=7
name="WSe2_"
```

```python
sp.load(datapath,name,run)
```

```python
#Visualize the data
def update(i=0):
    sp.axis[str(i)].plot()
i_slider = widgets.IntSlider(value=1,min=1,max=len(sp.axis),step=1,description='i')
interact(update, i=i_slider)
```

Tilt Scan (old version of files)

```python
datapath="test_data/"
fst=14 #first scan
last=35 #last scan
sp.load_tilt_old(datapath,fst,last)
```

```python
#Individual Visualization 
def update(i=1):
    sp.tilt[str(i)].plot()

i_slider = widgets.IntSlider(value=0,min=0,max=len(sp.tilt)-1,step=1,description='i')

interact(update, i=i_slider)
```

```python
# Combine Tilt scan in a single xarray
stacked = sp.combine_tilts() 
```
